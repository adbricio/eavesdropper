```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('graficas1_means.csv')
df1['mean_simulated_time'] = df1['mean_simulated_time'] * 10**(-9)
```

```python
df1_des = pd.read_csv('graficas1_results.csv')
std_dev1 = df1_des.groupby('percentage')['qber'].std().to_numpy()

for i in range(len(std_dev1)):
    std_dev1[i] = 1.96*std_dev1[i]/10
```

```python
std_dev1
```

```python
df2 = pd.read_csv('graficas2_means.csv')
df2['mean_simulated_time'] = df2['mean_simulated_time'] * 10**(-9)
```

```python
df2_des = pd.read_csv('graficas2_results.csv')

percs = df2_des['percentage'].unique()

std2 = {}

for perc in percs:
    std2[perc] = df2_des[df2_des['percentage']==perc].groupby('key_length')['qber'].std().to_numpy()
    for i in range(len(std2[perc])):
        std2[perc][i] = 1.96*std2[perc][i]/10
                 
std2
```

```python
plt.errorbar(df1['percentage'], df1['mean_qber'], yerr =std_dev1,  fmt = 'o-', color = colors[0], ecolor = error_colors[0], capsize=7)
plt.xlabel('Percentage of spied bits')
plt.ylabel('Mean QBER')
plt.title('256-bit key')
```

Como es lógico, cuantos más bits espía el eavesdropper más QBER hay. 

- Cuando espía 0 bits el QBER es de 0 porque la simulación está sin ruido, así que todo sale perfecto
- Cuando espía el 99% de los bits el QBER es cercano a 0.5. Esto creo que es porque si Eve espía todos los bits es como si el proceso fuese completamente aleatorio, entonces como bob puede elegir o 0 o 1 y Alice igual, la probabilidad de que coincidan es del 50%

- Vale es que creo que el QBER siempre es aprox la mitad de los bits espiados, pensar esto por que es

```python
percs = df2['percentage'].unique()

for (i,perc) in enumerate(percs):
    plt.errorbar(df2[df2['percentage'] == perc]['key_length'], df2[df2['percentage'] == perc]['mean_qber'], fmt = '-o', yerr = std2[perc], color = colors[i],ecolor = error_colors[i], capsize=7, label = f'{perc} %')
plt.xlabel('Key size')
plt.ylabel('Mean QBER')
plt.title('Fixed percentage of spied bits')
plt.legend(loc=(0.8, 0.15))

plt.show()
```

El porcentaje de bits que se espían es mucho más significativo en el QBER que el tamaño de la clave. Con el mismo porcentaje, el QBER se mantiene muy similar para todos los tamaños de clave considerados
